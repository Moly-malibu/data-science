```python
result = model.query(scaler.transform(df[df['track_id'] == '7pD6katrwWH7H3Xt8cIZS9'].drop(['artist_name', 'track_id', 'track_name', 'duration_ms', 'popularity'], axis = 1)), k = 11)
```

```python
results = df.iloc[result[1][0][0:11]].copy()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ax = results.plot.barh(x = 'track_name', y = ['acousticness', 'danceability', 'energy', 'valence', 'instrumentalness', 'liveness'], 
                  figsize = (12, 12), rot = 5, title = "I doubt this chart makes it to the app!", legend = 'reverse')
fig = ax.get_figure()
fig.savefig('similarity.png')
plt.show()
```

```python
ax = results[1:11].plot.scatter(x = 'distance', y = 'energy', c = 'danceability', cmap = 'coolwarm', title = 'Divergence of energy with distance from original song')
fig = ax.get_figure()
fig.savefig('energy.png')
plt.show()
```
