```python
clrs = {1: 'red', 2: 'orange', 3: 'yellow', 4: 'green', 5: 'blue', 6: 'violet', 7: 'gray'}
```

```python
for i in range(1, 8):
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    labels = ['acousticness', 'danceability', 'energy', 'valence', 'instrumentalness', 'liveness']
    values = results.iloc[i][['acousticness', 'danceability', 'energy', 'valence', 'instrumentalness', 'liveness']].tolist()
    values = [x for x in values]
    num_vals = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vals, endpoint = False).tolist
    angles = list(angles())
    values += values[:1]
    angles += angles[:1]
    ax.plot(angles, values, color=clrs[i], linewidth=1, label = results.iloc[i]['track_name'])
    ax.fill(angles, values, color=clrs[i], alpha = .25)

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_thetagrids(np.degrees(angles), labels)
    ax.legend(loc='upper right')

    for label, angle in zip(ax.get_xticklabels(), angles):
        if angle in (0, np.pi):
            label.set_horizontalalignment('center')
        elif 0 < angle < np.pi:
            label.set_horizontalalignment('left')
        else:
            label.set_horizontalalignment('right')

fig.savefig('radar.png')
```
